# **Purchase Data**

```python
#Importing Excel sheet

data = pd.read_excel("/content/drive/MyDrive/Machine Learning /LabSession1_Purchase.xlsx")
data
```

```python
#Removing Columns with NaN

df = data.dropna(axis=1)
df
```

```python
# A will be Candies,Mangoes,MilkPackets,Payments
# C will be payments.

A = df[["Candies (#)","Mangoes (Kg)","Milk Packets (#)"]]
A
```

```python
C = df[["Payment (Rs)"]]
C
#C.shape
```

```python
# Rank of A
A.rank()
```

```python
# Inv Of  A
A_Inv = np.linalg.pinv(A)
A_Inv
```

```python
# Ax=C ===> x= (A)-1 X C
# Use the Pseudo-inverse to calculate the model vector X for predicting the cost of the products available with the vendor.

x = np.matmul(A_Inv, C)
x
```

```python
# Mark all customers(in â€œPurchase Dataâ€ table)with paymentsabove Rs. 200 as RICH and others as POOR.
# Develop a classifier model to categorize customers into RICH or POOR class based on purchase behavior.

df['Category'] = np.where(df['Payment (Rs)'] > 200, 'Rich', 'Poor')
df[['Category']]
```

# **IRCTC Stock Price**

```python
#Importing Excel sheet

data_Stock = pd.read_excel("/content/drive/MyDrive/Machine Learning /LabSession1_Stock.xlsx")
data_Stock
```

```python
# Mean of 'Price' (Population Mean)

Mean = data_Stock['Price'].mean()
print("Mean :",Mean)
```

```python
# Variance of 'Price'

Variance = data_Stock['Price'].var()
print("Variance :",Variance)
```

```python
# Select the price data for all Wednesdays and calculate the sample mean. Compare the mean with the population mean and note your observations.

Wed = data_Stock[data_Stock['Day'] == 'Wed']

Mean_Wed = Wed['Price'].mean()
print("Mean of price data for all Wednesdays :",Mean_Wed)

print("Mean of price data for all Wednesdays is LESSER than Population mean(Price)")
```

```python
# Select the price data for the month of Apr and calculate the sample mean. Compare the mean with the population mean and note your observations.

Apr = data_Stock[data_Stock['Month'] == 'Apr']

Mean_Apr = Apr['Price'].mean()
print("Mean of price data for the month of Apr :",Mean_Apr)

print("Mean of price data for the month of Apr is MORE than Population mean(Price)")
```

```python
# From the Chg% (available in column I) find the probability of making a loss over the stock.
# Manipulate all of the values in a series by using the lambda function

Chg_Loss = data_Stock[data_Stock['Chg%'] < 0]

Count_Loss = len(Chg_Loss)
Count_Total = len(data_Stock)
Prob_Loss = (Count_Loss/Count_Total)
print("Probability of making a loss over the stock :",Prob_Loss)
```
